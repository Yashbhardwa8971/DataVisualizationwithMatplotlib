```python
nf = pd.read_csv('/content/drive/MyDrive/AlmaBetter Notebooks/Assignments/Chapter-2/netflix_titles.csv')
nf.head(5)
```

```python
nf.info()
```

```python
nf.isna().sum()
```

```python
nf['director'] = nf['director'].fillna('Unknown')
nf['cast'] = nf['cast'].fillna('Unknown')
nf['country'] = nf['country'].fillna('Unknown')
nf.isna().sum()
```

```python
nf.dropna(inplace=True)
nf.isna().any()
```

```python
nf.shape
```

```python
nf.drop(['show_id','description'], inplace = True, axis =1)
```

```python
nf.shape
```

```python
nf.drop_duplicates(inplace = True)
```

```python
nf.shape
```

```python
sns.countplot(data = nf, x = 'type')
plt.title('Count of Movies and TV Shows')
plt.show()
```

```python
nf['date_added'] = nf['date_added'].apply(lambda x: pd.to_datetime(x.strip(), format = '%B %d, %Y'))
nf['year_added'] = pd.DatetimeIndex(nf['date_added']).year
nf.head(5)
```

```python
gp = nf.groupby('year_added')['title'].count().reset_index()
gp
```

```python
sns.lineplot(data = gp, x = 'year_added', y = 'title')
plt.title('Number of Titles added each year')
plt.show()
```

```python
nf.columns
```

```python
lst = nf['listed_in'].apply(lambda x : x.split(','))
lst
```

```python
combined_lst = []
for i in lst:
  combined_lst = combined_lst + i
print(combined_lst)
```

```python
s = pd.Series(combined_lst)
s
```

```python
plt.figure(figsize=(10,6))
sns.countplot( x = s, order=s.value_counts().index)
plt.title('Count of Movies and TV Shows by genre')
plt.xticks(rotation = 90, fontsize = 8)
plt.show()
```

```python
sns.countplot(data = nf, x = 'rating', order=nf['rating'].value_counts().index)
plt.xticks(rotation = 90, fontsize = 12)
plt.show()
```
